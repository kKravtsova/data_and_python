### Data set 1 - Calorie burning


````python
import pandas as pd
url = "https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/Data%20Cleaning%20Data%20Sets/dirtydata.csv"
cal_df = pd.read_csv(url)
cal_df
````

````python
cal_df.info()
````

````python
cal_df_no_na = cal_df.dropna()
cal_df_no_na
````

````python
cal_df_no_na.describe()
````

````python
# 450 for duration looks like anomaly or mistake
cal_df_no_na = cal_df_no_na.loc[cal_df_no_na["Duration"] != 450]
cal_df_no_na
````

````python
def norm_date(df):
  if df["Date"].isdigit() == False:
    return "-".join(df["Date"].split("/"))
  else:
    return df["Date"][:4] + '-' + df["Date"][4:6] + '-' + df["Date"][6:]

cal_df_no_na["Date"] = cal_df_no_na.apply(norm_date, axis = 1)
cal_df_no_na
````

````python
from google.colab import files
cal_df_no_na.to_csv('calory_clean.csv')
files.download('calory_clean.csv')
````
